# Projeto 3 — Análise de Comportamento de Clientes

## Objetivo
Analisar padrões de compra e identificar perfil de clientes em um e-commerce fictício.

```python
import pandas as pd

df = pd.read_csv("dataset_clientes.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

receita_categoria = df.groupby("categoria")["valor_total"].sum()

receita_categoria.plot(kind="bar")
plt.title("Receita por Categoria")
plt.show()
```

```python
ticket_genero = df.groupby("genero")["valor_total"].mean()

ticket_genero.plot(kind="bar")
plt.title("Ticket Médio por Gênero")
plt.show()
```

# Principais Insights

- A categoria com maior faturamento total foi **Moda Feminina**, indicando maior concentração de compras nesse segmento;
- O ticket médio do público feminino apresentou valor superior ao masculino;
- A faixa etária entre 25 e 35 anos concentra maior volume de compras;
- Clientes de grandes capitais como São Paulo e Rio de Janeiro apresentam maior valor acumulado de compras;
- Existe potencial para estratégias de fidelização focadas nos clientes com maior ticket médio.
